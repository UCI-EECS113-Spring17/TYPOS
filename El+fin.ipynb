```python
#reminder that this module needs to be called to get current time
import datetime 
import json
import urllib.request
import time
from pytz import timezone
from PIL import ImageFont
from PIL import ImageDraw 


def time_converter(time):
    converted_time = datetime.datetime.fromtimestamp(
        int(time)
    ).strftime('%I:%M %p')
    return converted_time


def url_builder(city_id):
    user_api = 'a2dbbae99a4db58e631047474b9ae5b6'  # Obtain yours form: http://openweathermap.org/
    unit = 'imperial'  # For Fahrenheit use imperial, for Celsius use metric, and the default is Kelvin.
    api = 'http://api.openweathermap.org/data/2.5/weather?id='     # Search for your city ID here: http://bulk.openweathermap.org/sample/city.list.json.gz

    full_api_url = api + str(city_id) + '&mode=json&units=' + unit + '&APPID=' + user_api
    return full_api_url


def data_fetch(full_api_url):
    url = urllib.request.urlopen(full_api_url)
    output = url.read().decode('utf-8')
    raw_api_dict = json.loads(output)
    url.close()
    return raw_api_dict


def data_organizer(raw_api_dict):
    data = dict(
        city=raw_api_dict.get('name'),
        country=raw_api_dict.get('sys').get('country'),
        temp=raw_api_dict.get('main').get('temp'),
        temp_max=raw_api_dict.get('main').get('temp_max'),
        temp_min=raw_api_dict.get('main').get('temp_min'),
        humidity=raw_api_dict.get('main').get('humidity'),
        pressure=raw_api_dict.get('main').get('pressure'),
        sky=raw_api_dict['weather'][0]['main'],
        sunrise=time_converter(raw_api_dict.get('sys').get('sunrise')),
        sunset=time_converter(raw_api_dict.get('sys').get('sunset')),
        wind=raw_api_dict.get('wind').get('speed'),
        wind_deg=raw_api_dict.get('deg'),
        dt=time_converter(raw_api_dict.get('dt')),
        cloudiness=raw_api_dict.get('clouds').get('all')
    )
    return data


def data_output(data):
    m_symbol = '\xb0' + 'F'
    fmt = "%Y-%m-%d %H:%M:%S %Z%z"
    now_utc = datetime.datetime.now(timezone('UTC'))
    now_pacific = now_utc.astimezone(timezone('US/Pacific'))
    
    data1 = 'Date & Time: '+ str(now_pacific.strftime(fmt))
    data2 = '---------------------------------------'
    data3 = 'Current weather in: {}, {}:'.format(data['city'], data['country'])
    data4 = str(data['temp']) +  " " + m_symbol + " " + str(data['sky'])
    data5 = 'Max: {}, Min: {}'.format(data['temp_max'], data['temp_min'])
    data6 = ''
    data7 = 'Wind Speed: {}, Degree: {}'.format(data['wind'], data['wind_deg'])
    data8 = 'Humidity: {}'.format(data['humidity'])
    data9 = 'Cloud: {}'.format(data['cloudiness'])
    data10 = 'Pressure: {}'.format(data['pressure'])
    
    all_data = data1 + "\n" + data2 + "\n" + data3 + "\n" + data4 + "\n" + data5 + "\n" + data6 + "\n" + data7 + "\n" + data8 + "\n" + data9 + "\n" + data10  
    print(all_data)
    return all_data


if __name__ == '__main__':
    try:
        print_this = data_output(data_organizer(data_fetch(url_builder(5195010))))
        weather_img = Image.open("black.jpg")
        draw = ImageDraw.Draw(weather_img)
        font = ImageFont.truetype("arial.ttf", 30)
        #font = ImageFont.truetype(<font-file>, <font-size>)
        #draw.text((x, y),"Sample Text",(r,g,b))
        draw.text((1920/2-300,1080/2-250), print_this,(255,255,255),font=font)
        weather_img.save('weather.jpg')
    except IOError:
        print('no internet')
        
        
```

```python
hdmi_out = HDMI('out')

size = 1920, 1080
img_resized = weather_img.resize(size, Image.ANTIALIAS)
frame = np.array(img_resized.getdata(),np.uint8).reshape(img_resized.size[0], img_resized.size[1], 3)

print("starting...")
np_frame= (np.frombuffer(frame, dtype=np.uint8)).reshape(1080,1920,3)
hdmi_out.frame_raw(bytearray(np_frame.tobytes()))

hdmi_out.mode(4)
hdmi_out.start()

for i in range(len(photos)):
    response = requests.get(photos[i])
    img = Image.open(BytesIO(response.content))
    size = 1920, 1080
    img_resized = img.resize(size, Image.ANTIALIAS)
    draw1 = ImageDraw.Draw(img_resized)
    font1 = ImageFont.truetype("arial.ttf", 50)
    draw1.text((1920/2-300,1080-200), caption[i],(255,255,255),font=font)
    frame = np.array(img_resized.getdata(),np.uint8).reshape(img_resized.size[0], img_resized.size[1], 3)
    
    print("current i:", i)
    np_frame= (np.frombuffer(frame, dtype=np.uint8)).reshape(1080,1920,3)
    hdmi_out.frame_raw(bytearray(np_frame.tobytes()))
print("slideshow has ended")
```
